```python
import heapq
import time
from copy import deepcopy
```

```python
def move_left(A):
    B=deepcopy(A)
    li=[]

    #GETTING THE VALUE OF THE INDEX HAVING VALUES AS 0
    for i in range(len(A)):
        for j in range(len(A)):
            if A[i][j] == 0 :
                li.append(i)
                li.append(j)

    if li[1] - 1 < 0 :
        return None

    else:
        r , c = li[0] , li[1]
        B[r][c] = B[r][c-1]
        B[r][c-1] = 0

    return B

#ans = move_left(A)
#print(ans)
```

```python
#A = [[0,2,3],[4,5,6],[7,8,9]]

def move_right(A):
    B=deepcopy(A)
    li=[]

    #GETTING THE VALUE OF THE INDEX HAVING VALUES AS 0
    for i in range(len(A)):
        for j in range(len(A)):
            if A[i][j] == 0 :
                li.append(i)
                li.append(j)

    if li[1] + 1 >= len(A) :
        return None

    else:
        r , c = li[0] , li[1]
        B[r][c] = B[r][c+1]
        B[r][c+1] = 0

    return B

#ans = move_right(A)
#print(ans)
```

```python
#A = [[1,2,3],[0,5,6],[7,8,9]]

def move_up(A):
    B=deepcopy(A)
    li=[]

    #GETTING THE VALUE OF THE INDEX HAVING VALUES AS 0
    for i in range(len(A)):
        for j in range(len(A)):
            if A[i][j] == 0 :
                li.append(i)
                li.append(j)

    if li[0] - 1 < 0  :
        return None

    else:
        r , c = li[0] , li[1]
        B[r][c] = B[r-1][c]
        B[r-1][c] = 0

    return B

#ans = move_up(A)
#print(ans)
```

```python
#A = [[1,2,3],[0,5,6],[7,8,9]]

def move_down(A):
    B=deepcopy(A)
    li=[]

    #GETTING THE VALUE OF THE INDEX HAVING VALUES AS 0
    for i in range(len(A)):
        for j in range(len(A)):
            if A[i][j] == 0 :
                li.append(i)
                li.append(j)

    if li[0] + 1 >= len(A)  :
        return None

    else:
        r , c = li[0] , li[1]
        B[r][c] = B[r+1][c]
        B[r+1][c] = 0

    return B

#ans = move_down(A)
#print(ans)
```

```python
def find_path(parent):
    path=[]
    path.append(parent[len(parent)-1][1])
    path.append(parent[len(parent)-1][0])
    x=parent[len(parent)-1][0]
    for par in range(len(parent)-2,-1,-1):
        if parent[par][1]==x:
            x=parent[par][0]
            path.append(x)
    return path
#driver code
# x=find_path([['E','A'],['E','B'],['E','D'],['E','H'],['A','C']])
# print(x)
```

```python
def h1(start, goal):
    cnth1=0
    for i in range(len(start[0])):
        for j in range(len(start[0])):
            if start[i][j]!=goal[i][j] and start[i][j]!=0:
                cnth1+=1
    return cnth1
        
```

```python
def h2(li):
    initial_state = []
    for i in range(len(li)):
        for j in range(len(li[i])):
            initial_state.append(li[i][j])
  
    cnth2 = sum(abs(((val-1)%3 - i%3) + abs((val-1)//3 - i//3))
    for i,val in enumerate(initial_state) if val)
    return cnth2

#driver code
state = [[1,5,3],[4,2,6],[7,8,0]]
print(h2(state))
```

```python
def bfs_goal_search(start,goal):
    explored=[]
    queue=[]
    new_queue = [start]
    child=[]
    parent=[]
    heapq.heappush(queue,[0,0,start])
    if(start==goal):
        return print("Start is goal")
    count=0
    while queue:
        count+=1
        print(count)
        node1 = heapq.heappop(queue)
        node = node1[-1]
        print(node1)
        print("count",count)
        g=node1[1]
        if node not in explored:
            explored.append(node)
        
        #left
        child = move_left(node)
        if child!=None:
            if child not in new_queue and child not in explored:
                g1=g+1
                h=h1(child,goal)
                heapq.heappush(queue,[g1+h,g1,child])
                new_queue.append(child)
            if child == goal:
                #return find_path(parent)
                return child
        
        #right
        child = move_right(node)
        if child!=None:
            if child not in new_queue and child not in explored:
                g1=g+1
                h=h1(child,goal)
                heapq.heappush(queue,[g1+h,g1,child])
                new_queue.append(child)
            if child == goal:
                #return find_path(parent)
                return child
        
        #up
        child = move_up(node)
        if child!=None:
            if child not in new_queue and child not in explored:
                g1=g+1
                h=h1(child,goal)
                heapq.heappush(queue,[g1+h,g1,child])
                new_queue.append(child)
            if child == goal:
                #return find_path(parent)
                return child
        
        #down
        child = move_down(node)
        if child!=None:
            if child not in new_queue and child not in explored:
                g1=g+1
                h=h1(child,goal)
                heapq.heappush(queue,[g1+h,g1,child])
                new_queue.append(child)
            if child == goal:
                #return find_path(parent)
                return child
            
start=time.time()
start1=[[4,5,7],[3,1,6],[8,2,0]]            
goal1=[[1,2,3],[4,5,6],[7,8,0]]
bfs_goal_search(start1,goal1)
end=time.time()
print("Total Time : " ,end-start)
```
