```python
# import required modules
import requests

page = requests.get("https://en.wikipedia.org/wiki/Computer")
print(page.content)
```

```python

from bs4 import BeautifulSoup


soup = BeautifulSoup(page.content, 'html.parser')

#  scrapped data
print(soup.prettify())
```

```python
# PARAGRAPH

list(soup.children)

# print(soup.find_all('p'))

print('\n\n')

for i in soup.find_all('p'):
    print(i.get_text())
    print()
# print(soup.find_all('p')[1].get_text())
```
